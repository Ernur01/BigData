```python
import math
x = np.linspace(0,math.pi*2,100)
y = np.sin(x) + np.random.rand(100)*0.2
plt.scatter(x,y,s=15,c='brown')
```

```python
x = np.linspace(0,math.pi*2,100)
y = np.sin(x) + np.random.rand(100)*0.2
plt.subplot(1,2,1)
plt.scatter(x,y,s=5)

x = np.linspace(0,math.pi*4,100)
y = np.sin(x)
plt.subplot(1,2,2)
plt.plot(x,y, "r+-")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.4,0.3])
axes1.plot(x,y,'b')
axes2.plot(x,y,'r')
axes2.set_xlabel('dada')
axes2.set_title('11111dada')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
for row in axes:
    for ax in row:
        ax.scatter(np.random.rand(10),np.random.rand(10))
        ax.set_title('dadada')
plt.tight_layout()
plt.savefig('ttt.png')
```
