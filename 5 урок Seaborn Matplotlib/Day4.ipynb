```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,"b")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,.4,0.3])

axes1.plot(x,y,"b")
axes2.plot(y,x,"g")
axes2.set_xlabel("X label")
axes2.set_title("my title")
```
